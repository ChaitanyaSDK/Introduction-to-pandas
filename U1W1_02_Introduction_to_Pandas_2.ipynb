### b) We have some made-up data for this exercise. It is in the form of a dictionary. The field *area* has to be used as the index. Create a Dataframe using the data.  
 
 
 


```
data = {'Capitals': ['PANAJI', 'CHANDIGARH', 'PATNA', 'AGARTALA'],

      'Population': ['1,347,668', '21,144,564', '82,998,509','3,199,203'],
 
      'States': ['GOA', 'HARYANA', 'BIHAR', 'TRIPURA'],
 
      'Visitors': [300, 4000, 50, 80000],

      'area' : ["3,702", "44,212" ,"94,163", "10,486"]}
```

You can refer the below link to know about the syntax to create a dataframe 


https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html

```
# Your Code Here

df=pd.DataFrame(data=data)
print(df)
```

### c) Display only Capitals and States columns from the dataframe

```
# Your Code Here
print(df[['Capitals','States']])
```

### d) What is the total number of visitors?

```
# Your Code Here
print(df['Visitors'].sum())
```
